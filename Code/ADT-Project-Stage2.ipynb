### Adding data to SQL


```python
import sqlite3  
import pandas as pd
```

```python
conn = sqlite3.connect('project.db')
c = conn.cursor()
```

```python
app_data = pd.read_csv(r'sampled_appstore.csv')
```

```python
# Author - Anushree
# Definng schema for the applications table
create_table_query = """
CREATE TABLE applications (
    App_Id INT PRIMARY KEY,
    App_name VARCHAR(255) NOT NULL,
    Developer_Id INT,
    Genre VARCHAR(255) NOT NULL,
    Size FLOAT,
    App_version VARCHAR(20),
    IOS_version VARCHAR(20),
    Released_date DATE,
    Updated_date DATE,
    Avg_user_rating FLOAT,
    Age_group VARCHAR(10),
    FOREIGN KEY (Developer_Id) REFERENCES developers(Developer_Id)
);
"""

c.execute(create_table_query)
conn.commit()



# Defining the schema for the developers table
create_table_query = """
CREATE TABLE developers (
    Developer_Id INT PRIMARY KEY,
    Developer VARCHAR(255) NOT NULL
);
"""

c.execute(create_table_query)
conn.commit()



# Defining the schema for the pricing table
create_table_query = """
CREATE TABLE pricing (
    App_Id INT PRIMARY KEY,
    Price DECIMAL(10, 2) NOT NULL,
    Currency VARCHAR(10) NOT NULL,
    FOREIGN KEY (App_Id) REFERENCES applications(App_Id)
);
"""

c.execute(create_table_query)
conn.commit()
```

```python
# Author - Anushree
#Inserting data in the applications table

insert_query = """
INSERT INTO applications (App_Id, App_name, Developer_Id, Genre, Size, App_version, IOS_version, Released_date, Updated_date, Avg_user_rating, Age_group)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
"""

for index, row in app_data.iterrows():
    values = (
        row['App_Id'], row['App_name'], row['Developer_Id'], row['Genre'], row['Size'],
        row['App_version'], row['IOS_version'], row['Released_date'], row['Updated_date'],
        row['Avg_user_rating'], row['Age_group']
    )

    c.execute(insert_query, values)

conn.commit()



# Inserting data in the developers table

unique_developer_ids = set()

# Iterate through the DataFrame to collect unique developer IDs
for index, row in app_data.iterrows():
    developer_id = row['Developer_Id']
    developer = row['Developer']
    if developer_id not in unique_developer_ids:
        unique_developer_ids.add(developer_id)
        insert_query = """
            INSERT INTO developers (Developer_Id, Developer)
            VALUES (?, ?);
            """
        values = (developer_id, developer)
        c.execute(insert_query, values)

conn.commit()



# Inserting data in the pricing table
for index, row in app_data.iterrows():
    insert_query = """
    INSERT INTO pricing (App_Id, Price, Currency)
    VALUES (?, ?, ?)
    """
    values = (row['App_Id'], row['Price'], row['Currency'])
    
    c.execute(insert_query, values)


conn.commit()
```

### Views

```python
# Author - Anushree, Shubhangi
# Top Developers View: top developers by the number of applications they've developed

c.execute("""CREATE VIEW TopDevelopers AS
SELECT d.developer, COUNT(a.app_id) AS total_apps
FROM developers d
LEFT JOIN applications a ON d.developer_id = a.developer_id
GROUP BY d.developer_id, d.developer
ORDER BY total_apps DESC;
 """)
```

```python
c.execute("""SELECT * 
FROM TopDevelopers
LIMIT 10;
""")
c.fetchall()
```

```python
# Author - Anushree, Kaushik
# Free Applications View: list of free applications

c.execute("""CREATE VIEW FreeApplications AS
SELECT a.app_name, d.developer
FROM applications a
JOIN developers d ON a.developer_id = d.developer_id
JOIN pricing p ON a.app_id = p.app_id
WHERE p.price = 0;

 """)
```

```python
c.execute("""SELECT * 
FROM FreeApplications
LIMIT 15;
""")
c.fetchall()
```

```python
# Author - Anushree
# TopRated Applications by Genre View: apps with the highest user ratings based on genre

c.execute("""CREATE VIEW TopRatedAppsByGenre AS
SELECT a.genre, a.app_name, a.avg_user_rating
FROM applications a
WHERE a.avg_user_rating IS NOT NULL
AND a.avg_user_rating > 3
ORDER BY a.avg_user_rating DESC;
 """)
```

```python
c.execute("""SELECT * 
FROM TopRatedAppsByGenre as t
WHERE t.genre = 'Book'
LIMIT 10;
""")
c.fetchall()
```
